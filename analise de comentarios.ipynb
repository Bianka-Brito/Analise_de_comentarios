```python
import pandas as pd
```

```python
df= pd.read_excel('comentarios.xlsx') 
```

```python
#função que separa os comentarios 
def analise(comentarios):
    comentarios=comentarios.lower()
    elogios= ['otimo', 'excelente', 'bom', 'legal', 'incríveis', 'incrível', 'dispostas', 'ótima', 'gostei', 'orgulho', 'respeito', 'autonomia', 'liberdade']
    criticas= ['ruim', 'péssimo', 'não gostei', 'falha', 'falta', 'problema', 'crítica', 'precisa melhorar']

#any: se pelo menos uma opção for verdadeira 
    if any( palavras in comentarios for palavras in elogios): 
        return 'elogios'
    elif any( palavras in comentarios for palavras in criticas):
        return 'criticas'
    else: 
        return 'outros'

#seleciona a coluna resposta e aplica a função em todos os valores de cada linha 
df['classificação'] = df['Respostas'].apply(analise) 

#cria novos df classificando pelo filtro 
criticas= df[df['classificação'] == 'criticas'] 
elogios= df[df['classificação'] == 'elogios']
outros= df[df['classificação'] == 'outros']

#save
criticas.to_excel('testCriticas.xlsx')
elogios.to_excel('testElogios.xlsx')
outros.to_excel('testOutros.xlsx')
```
