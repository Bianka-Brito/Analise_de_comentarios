```python
import pandas as pd 
import ollama 
```

```python
df=pd.read_excel('comentarios.xlsx')
texto=df.to_string(index=False)
```

```python
prompt=f"Analise os seguintes comentários extraídos de uma planilha:\n\n{texto}\n\nClassifique o clima geral como positivo, negativo ou neutro. Responda em português."
resposta=ollama.generate(model='llama3',prompt=prompt)
print(resposta['response'])
```

```python
prompt=f"""mostre elogios originais do excel em portugues e dividido por topicos

{texto}
"""
resposta=ollama.generate(model='llama3', prompt=prompt)
print(resposta['response'])
```

```python
prompt=f"""
Analise os comentarios e agrupe-os por topicos. Para cada topico siga este formato: 
[Titulo do topico]: [quantidade de comentarios por topicos] menções.
[Texto analitico em PORTUGUES BR]

comentarios:
{texto}
"""
resposta=ollama.generate(model='llama3', prompt=prompt)
print(resposta['response'])
```
